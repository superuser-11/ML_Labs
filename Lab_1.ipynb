```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#x_train : this is the input variable contain size of house in Sqfeet (size is 1000sqfeet)
#y_tarin : this is output variable contain target values, which is the price of houses in (1000s $)
x_train = np.array([1.0 , 2.0])
y_train = np.array([300.0 , 500.0])

#m is the number of examples or observations
m=len(x_train)
print("number of examples are: " , m)
```

```python
i=0 #change this to 1 to see x_1 and y_1
x_i=x_train[i]
y_i=y_train[i]
# print(f"(x^({i}) , y^({i})) = ({x_i} , {y_i})")
print(f"(X^({i}) , Y^({i})) = ({x_i} , {y_i})")
```

```python
#plotting the data in the graph
plt.scatter(x_train , y_train , marker='x' , c='r' , label='actual')
#set the title
plt.title("Housing Prices")
#set the y-axis label
plt.ylabel('Price (in 1000s of dollars)')
#set the x-axis label
plt.xlabel('Size (1000 sqft)')
plt.show()
```

```python
# now we will predict the values for the same inputs
# picking random values for w and b
w=100 #set to 200 to see a better result
b=100
print(f"w: {w}")
print(f"b: {b}")
```

```python
#predicting values for the inputs
def calculate_model_output(w , b , x):
   f_wb = np.zeros(len(x))
   for i in range(len(x)):
      f_wb[i] = w * x[i] + b

   return f_wb

# function calling
temp_f_wb= calculate_model_output(w , b , x_train)

#plot the model prediction
plt.plot(x_train , temp_f_wb , c='b' , label="our prediction")
#plot the data point
plt.scatter(x_train , y_train , marker='x' , c='r' , label="Actual values")
plt.title("Housing Prices")
plt.ylabel('Price (in 1000s of dollars)')
plt.xlabel('Size (1000 sqft)')
plt.legend()
plt.show()
```
