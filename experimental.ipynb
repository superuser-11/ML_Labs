```python
#bad model
import numpy as np
import matplotlib.pyplot as plt
```

```python
#x_train : this is the input variable contain size of house in Sqfeet (size is 1000sqfeet)
#y_tarin : this is output variable contain target values, which is the price of houses in (1000s $)
x_train = np.array([1000 , 2000 , 3000])
y_train = np.array([190 , 310 , 500])
m=len(x_train)
print(f"number of training examples is : {m}")


```

```python
# now we will predict the values for the same inputs
# picking random values for w and b
w=0.06
b=50
print(f"w: {w}")
print(f"b: {b}")
```

```python
#predicting values for the inputs
def calculate_model_output(w , b , x):
   f_wb = np.zeros(len(x))
   for i in range(len(x)):
      f_wb[i] = w * x[i] + b
      print(f"y^{i} = {f_wb[i]}") #to print the predicted values

   return f_wb

# function calling
temp_f_wb= calculate_model_output(w , b , x_train)

#plot the model prediction
plt.plot(x_train , temp_f_wb , c='g' , label="our prediction")
#plot the data point
plt.scatter(x_train , y_train , marker='X' , c='r' , label="Actual values")
plt.title("Housing Prices")
plt.ylabel('Price (in 1000s of dollars)')
plt.xlabel('Size (1000 sqft)')
plt.legend()
plt.show()
```

```python
#Using cost function J(w , b) to calculate the distance between y-actual and y^ - predicted
cost_f_wb = np.zeros(len(x_train))
for i in range(len(x_train)):
      cost_f_wb[i] = pow(temp_f_wb[i] - y_train[i] , 2)  #pow is used to take square
      print(f"y^{i} - y{i} = {cost_f_wb[i]}") # distance of y^-y for each example

cost = (np.sum(cost_f_wb))  #np.sum is used to add the values in the array
print(f"cost = {cost}")

print(f"J{w , b} = {cost / (2*m) }") #total cost after dividing with 1/2m
```
